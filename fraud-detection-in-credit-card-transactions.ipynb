```python
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
df=pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Class'].value_counts()
```

```python
data=pd.concat([df[df['Class']==1],df[df['Class']==0].head(492)],axis=0)
```

```python
data['Class'].value_counts()
```

```python
data.duplicated().sum()
```

```python
data=data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
x=data.drop('Class',axis=1)
y=data['Class']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x_standarised=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_standarised,y,test_size=.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
gc=GridSearchCV(knn,[{'n_neighbors':[1,4,7,10]}])
```

```python
gc.fit(x_train,y_train)
```

```python
gc.best_params_
```

```python
knn=gc.best_estimator_
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
sns.heatmap((confusion_matrix(y_test,knn.predict(x_test))),annot=True,fmt='d')
```
